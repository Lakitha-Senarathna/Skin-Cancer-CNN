```python
from PIL import Image
import os

input_dir = r"D:\# University of Colombo  FOS\Semester 6\PH 3022 - Machine Learning and Neural Computation\skin_cancer_dataset\Benign_reduced"
output_dir = r"D:\# University of Colombo  FOS\Semester 6\PH 3022 - Machine Learning and Neural Computation\skin_cancer_dataset\Resized\Benign"
os.makedirs(output_dir, exist_ok=True)

for img_name in os.listdir(input_dir):
    img = Image.open(os.path.join(input_dir, img_name))
    img = img.resize((224, 224))
    img.save(os.path.join(output_dir, img_name), optimize=True, quality=85)
```
